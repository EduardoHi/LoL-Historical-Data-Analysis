En nuestros datos, el objeto principal es la clase `MatchInfo`, este contiene información general de las partidas, asi como tambien la informacion de cada equipo, y la informacion de los eventos que ocurrieron en la partida: Asesinatos, Destruir Estructuras, y Conseguir objetivos neutrales(mounstros)

Otro aspecto interesante es que los `ENUMS` son _ortogonales e independientes_ entre ellos, entonces cada uno representa efectivamente un punto de división de los datos. por poner un ejemplos, `Year` divide totalmente el conjunto de los elementos `Match`, igual con `Season`. Mientras que `Lane` divide totalmente las estructuras. And so on.
